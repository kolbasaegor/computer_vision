```python
import cv2 as cv
import numpy as np
from PIL import Image

cap = cv.VideoCapture(1) #захват видео
color_name = ['yellow','red','blue','green']
karlik = [0,0,0,0]
fed_color = False
dolg =50 # сколько нунжно кадров для определения цвета

while True: #пока не определился цвет
    ret, frame = cap.read()
    frameCopy = frame.copy()
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    hsv = cv.blur(hsv, (3,3))
      
    #cv.imshow('binary_hsv', mask)
    
    if not fed_color:
        
        mask = cv.inRange(hsv, (0 ,122, 37), (187, 249, 255)) #бинаризую картинку 
        mask = cv.erode(mask, None, iterations = 1)
        mask = cv.dilate(mask, None, iterations = 2)
        
        contours0, hierarchy = cv.findContours( mask.copy(), cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
        if contours0:


            contours = sorted(contours0, key = cv.contourArea, reverse = True) #если нашлись контуры, то сортирую
                                                                               # контуры по размеру области 
            cnt = contours[0]
            (x,y,w,h) = cv.boundingRect(cnt)
            if w>150 and h>150:
                rect = cv.minAreaRect(cnt) # пытаемся вписать прямоугольник
                box = cv.boxPoints(rect) # поиск четырех вершин прямоугольника
                box = np.int0(box) # округление координат
                cv.drawContours(frame,[box],0,(255,0,255),2) # рисуем прямоугольник

                roImg = frameCopy[y:y+h, x:x+w] # обрезаю контур
                roImg = cv.resize(roImg, (128,128))
                cv.imshow('obrez', roImg)

                color_pix = [0,0,0,0]


                colorset =[
                [15, 160, 215], #yellow
                [20,25,160], #red
                [150, 43, 25], #blue
                [30, 125, 15] #green
                ]
                min_razn = 50000
                zzz = 0

                    #попиксильное сравнивание и выделение цвета

                for i in range(0,len(roImg),4):
                    for k in range(0,len(roImg[i]),4):
                        for j in range(4):
                            razn = abs(roImg[i][k][0] - colorset[j][0])+abs(roImg[i][k][1] - colorset[j][1])+abs(roImg[i][k][2] - colorset[j][2])
                            if razn< min_razn:
                                min_razn = razn 
                                zzz = j
                        color_pix[zzz] += 1     

                max_color = 0
                mxdex = 0
                for i in range(4):
                    if color_pix[i] > max_color:
                        max_color = color_pix[i]
                        mxdex = i
                if max_color > 500:            
                    cv.putText(frame, color_name[mxdex], (20, 20), cv.FONT_HERSHEY_SIMPLEX, 1, colorset[mxdex], 2) #вывод надписи цвета
                    karlik[mxdex] += 1
                    for k in range(len(karlik)):
                        if karlik[k] > dolg:
                            fed_color = True
                            break
                else:
                    cv.putText(frame, 'IDK', (20, 20), cv.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 2) #вывод надписи нецвета
        
    else:
        cv.putText(frame, 'find: '+color_name[k], (20, 20), cv.FONT_HERSHEY_SIMPLEX, 1, colorset[k], 2) #вывод надписи нецвета
    


    cv.imshow('contours', frame) # вывод обработанного кадра в окно
        
    if cv.waitKey(1) == 27:
        break

cap.release()
cv.destroyAllWindows()
```
